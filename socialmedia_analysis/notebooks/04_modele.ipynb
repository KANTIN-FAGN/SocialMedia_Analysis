```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../data/data_cleaning.csv")
```

```python
X = df[["Daily_Screen_Time(hrs)"]]
y = df["Stress_Level(1-10)"]

model = LinearRegression()
model.fit(X, y)

print("Coefficient :", model.coef_[0])
print("Intercept :", model.intercept_)
```

```python
plt.scatter(df["Daily_Screen_Time(hrs)"], df["Stress_Level(1-10)"])
plt.plot(df["Daily_Screen_Time(hrs)"], model.predict(X), color="red")
plt.show()
```
