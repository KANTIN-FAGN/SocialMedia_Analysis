```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Mental_Health_and_Social_Media_Balance_Dataset.csv")
df.head()
```

```python
df.drop("User_ID", axis=1, inplace=True)
df.drop("Age", axis=1, inplace=True)
df
```

```python
df["Gender"].unique()
```

```python
df["Social_Media_Platform"].unique()
```

```python
df = pd.get_dummies(df, columns=["Gender", "Social_Media_Platform"])
df.head()
```

```python
bins = [0, 2, 4, 6, 24]
labels = ["0-2h", "2-4h", "4-6h", "6h+"]

df["usage_group"] = pd.cut(df["Daily_Screen_Time(hrs)"], bins=bins, labels=labels)
df.head()
```

```python
df.to_csv('../data/data_cleaning.csv', index=False)
```
